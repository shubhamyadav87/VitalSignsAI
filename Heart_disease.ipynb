```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv("/content/heart_disease.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data["dataset"].value_counts()
```

```python
data.drop(columns=["id","thal"],inplace=True)
```

```python
data.info()
```

```python
columns_encode=["dataset","sex","cp","fbs","restecg","exang","slope"]
for columns in columns_encode:
  data[columns]=LabelEncoder().fit_transform(data[columns])
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.head()
```

```python
columns_to_int=["trestbps","chol","thalch","oldpeak","ca"]
data[columns_to_int].astype(int)
```

```python
data.info()
```

```python
data["target"].isna().sum()
```

```python
data["target"]=data["target"].astype(int).replace([1,2,3,4],1)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
x=data.drop(columns=["target"],axis=1)
y=data["target"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model2=RandomForestClassifier()
```

```python
model2.fit(x_train,y_train)
```

```python
y_pred=model2.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
import pickle
```

```python
filename="Heart_disease_prediction.sav"
pickle.dump(model2,open(filename,"wb"))
```
